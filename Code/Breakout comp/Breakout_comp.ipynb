```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Plotting
interval = 1000

reward_hist_no_exp = pickle.load(open('all_reward_no_exp.p','rb'))
# reward_hist_narrow = pickle.load(open('all_reward_narrow_dropout.p','rb'))
reward_hist_dropout = pickle.load(open('all_reward_dropout.p','rb'))
# reward_hist_wide = pickle.load(open('all_reward_wide_dropout.p','rb'))
reward_hist_narrow = pickle.load(open('all_reward_narrow.p','rb'))
# reward_hist_cnn = pickle.load(open('all_reward_cnn.p','rb'))

def get_mean_var(reward_hist):
    ts = pd.Series(reward_hist[:20000])
    return pd.rolling_mean(ts, interval), pd.rolling_var(ts, interval)

reward_hist_no_exp_mean, reward_hist_no_exp_var = get_mean_var(reward_hist_no_exp)
# reward_hist_narrow_mean, reward_hist_narrow_var = get_mean_var(reward_hist_narrow)
# reward_hist_wide_mean, reward_hist_wide_var = get_mean_var(reward_hist_wide)
reward_hist_dropout_mean, reward_hist_dropout_var = get_mean_var(reward_hist_dropout)
reward_hist_narrow_mean, reward_hist_narrow_var = get_mean_var(reward_hist_narrow)
# reward_hist_cnn_mean, reward_hist_cnn_var = get_mean_var(reward_hist_cnn)

fig,axe = plt.subplots(2,1,figsize=(8,8))
plt.suptitle('PG Breakout Comparison plot', fontsize=20)
axe[0].set_ylim(1,8)
axe[0].plot([7]*20000,'b--')
axe[0].plot(reward_hist_no_exp_mean,label='No exploration')
# axe[0].plot(reward_hist_narrow_mean,label='Narrow network')
# axe[0].plot(reward_hist_wide_mean,label='Wide network')
axe[0].plot(reward_hist_dropout_mean,label='Dropout')
axe[0].plot(reward_hist_narrow_mean,label='Narrow network')
# axe[0].plot(reward_hist_cnn_mean,label='CNN')
axe[0].legend(loc=4,fontsize=10) 
axe[0].set_title('Mean',fontsize=15)
axe[1].set_ylim(0,16)
axe[1].plot(reward_hist_no_exp_var,label='No exploration')
# axe[1].plot(reward_hist_narrow_var,label='Narrow network')
# axe[1].plot(reward_hist_wide_var,label='Wide network')
axe[1].plot(reward_hist_dropout_var,label='Drop out')
axe[1].plot(reward_hist_narrow_var,label='Narrow networkt')
# axe[1].plot(reward_hist_cnn_var,label='CNN')
axe[1].legend(loc=4,fontsize=10) 
axe[1].set_title('Variance',fontsize=15)
plt.show()
fig.savefig('Breakout_comp.jpg')
```
