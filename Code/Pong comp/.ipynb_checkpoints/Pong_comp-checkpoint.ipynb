```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

interval =500

reward_hist_no_exp = pickle.load(open('all_reward_no_exp.p','rb'))
reward_hist_wide = pickle.load(open('all_reward_wide.p','rb'))
reward_hist_eps = pickle.load(open('all_reward_eps.p','rb'))
reward_hist_dropout = pickle.load(open('all_reward_dropout.p','rb'))
# reward_hist_cnn_wide = pickle.load(open('all_reward_cnn_wide_filter.p','rb'))
# reward_hist_wide_nodropout = pickle.load(open('all_reward_wide_nodropout.p','rb'))

def get_mean_var(reward_hist):
    ts = pd.Series(reward_hist[:10000])
    return pd.rolling_mean(ts, interval), pd.rolling_var(ts, interval)

reward_hist_no_exp_mean, reward_hist_no_exp_var = get_mean_var(reward_hist_no_exp)
reward_hist_wide_mean, reward_hist_wide_var = get_mean_var(reward_hist_wide)
reward_hist_eps_mean, reward_hist_eps_var = get_mean_var(reward_hist_eps)
reward_hist_dropout_mean, reward_hist_dropout_var = get_mean_var(reward_hist_dropout)
# reward_hist_cnn_wide_mean, reward_hist_cnn_wide_var = get_mean_var(reward_hist_cnn_wide)
# reward_hist_wide_nodropout_mean, reward_hist_wide_nodropout_var = get_mean_var(reward_hist_wide_nodropout)

fig,axe = plt.subplots(2,1,figsize=(7,7))
plt.suptitle('PG Pong Comparison plot', fontsize=20)
axe[0].set_ylim(-22,10)
axe[0].plot([5]*10000,'b--',label='Reward=5')
axe[0].plot(reward_hist_no_exp_mean,label='No exploration')
axe[0].plot(reward_hist_wide_mean,label='Wider network')
axe[0].plot(reward_hist_eps_mean,label='Epsilon greedy')
axe[0].plot(reward_hist_dropout_mean,label='Dropout')
# axe[0].plot(reward_hist_cnn_wide_mean,label='Wide CNN')
# axe[0].plot(reward_hist_wide_nodropout_mean,label='Wide wo dropout')
axe[0].legend(loc=4,fontsize=8) 
axe[0].set_title('Mean',fontsize=15)
axe[1].plot([35]*10000,'b--',label='Variance=35')
axe[1].plot(reward_hist_no_exp_var,label='No exploration')
axe[1].plot(reward_hist_wide_var,label='Wider network')
axe[1].plot(reward_hist_eps_var,label='Epsilon greedy')
axe[1].plot(reward_hist_dropout_var,label='Drop out')
# axe[1].plot(reward_hist_cnn_wide_var,label='Wide CNN')
# axe[1].plot(reward_hist_wide_nodropout_var,label='Wide wo dropout')
axe[1].legend(loc=4,fontsize=8) 
axe[1].set_title('Variance',fontsize=15)
plt.subplots_adjust(hspace=0.2)
plt.show()
fig.savefig('Pong_comp.jpg')
```
