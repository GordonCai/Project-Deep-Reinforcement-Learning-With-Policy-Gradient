```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

interval =500

reward_hist_no_exp = pickle.load(open('all_reward_no_exp.p','rb'))
reward_hist_wide = pickle.load(open('all_reward_wide.p','rb'))
reward_hist_eps = pickle.load(open('all_reward_eps.p','rb'))
reward_hist_dropout = pickle.load(open('all_reward_dropout.p','rb'))

def get_mean_var(reward_hist):
    ts = pd.Series(reward_hist[:10000])
    return pd.rolling_mean(ts, interval), pd.rolling_var(ts, interval)

reward_hist_no_exp_mean, reward_hist_no_exp_var = get_mean_var(reward_hist_no_exp)
reward_hist_wide_mean, reward_hist_wide_var = get_mean_var(reward_hist_wide)
reward_hist_eps_mean, reward_hist_eps_var = get_mean_var(reward_hist_eps)
reward_hist_dropout_mean, reward_hist_dropout_var = get_mean_var(reward_hist_dropout)

fig,axe = plt.subplots(2,1,figsize=(6,6))
plt.suptitle('Policy Gradient Pong', fontsize=18)
axe[0].set_ylim(-22,10)
axe[0].plot([5]*10000,'b--')
axe[0].plot(reward_hist_no_exp_mean,label='ANN')
axe[0].plot(reward_hist_wide_mean,label='Wider ANN')
axe[0].plot(reward_hist_eps_mean,label='CNN')
axe[0].plot(reward_hist_dropout_mean,label='Dropout')
axe[0].set_xlabel('Episode')
axe[0].set_ylabel('Rewards')
axe[0].legend(loc=4,fontsize=10) 
axe[0].set_title('Running Mean',fontsize=15)
axe[1].plot(reward_hist_no_exp_var,label='ANN')
axe[1].plot(reward_hist_wide_var,label='Wider ANN')
axe[1].plot(reward_hist_eps_var,label='CNN')
axe[1].plot(reward_hist_dropout_var,label='Drop out')
axe[1].set_xlabel('Episode')
axe[1].legend(loc=4,fontsize=10) 
axe[1].set_title('Variance',fontsize=15)
plt.subplots_adjust(hspace=0.4)
plt.show()
fig.savefig('Pong_comp.jpg')
```
